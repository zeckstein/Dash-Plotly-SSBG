```python
import plotly.express as px
import pandas as pd
import numpy as np

# Create sample time series data
np.random.seed(42)
dates = pd.date_range(start="2024-01-01", periods=30)
data = pd.DataFrame({
    "Date": dates,
    "Series A": np.cumsum(np.random.randn(30)),
    "Series B": np.cumsum(np.random.randn(30)),
    "Series C": np.cumsum(np.random.randn(30))
})

# Melt data for Plotly Express
df_melted = data.melt(id_vars="Date", var_name="Series", value_name="Value")

# Create interactive line chart
fig = px.line(
    df_melted,
    x="Date",
    y="Value",
    color="Series",
    title="Interactive Line Graph with Enhanced Legend"
)

# Improve legend interaction & layout
fig.update_layout(
    legend=dict(
        title="Click to Toggle Series",
        orientation="h",          # Horizontal legend
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    ),
    hovermode="x unified",        # Single hover label for all series
    legend_itemclick="toggle",    # Click legend to hide/show
    legend_itemdoubleclick="toggleothers"  # Double-click to isolate
)

# Optional: Highlight traces on hover
fig.update_traces(mode="lines+markers", hovertemplate="%{y:.2f}")

# Show the figure
fig.show()
```
